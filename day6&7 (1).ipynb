```python
class node:
  def __init__(self,data):
    self.data=data
    self.next=None
    self.prev=None
class DoublyLinkedList:
  def __init__(self):
    self.head=None
  def insertAtHead(self,data):
    if self.head is None:
      new=node(data)
      self.head=new
    else:
      new=node(data)
      new.next=self.head
      self.head.prev=new
      self.head=new

  def insertAtEnd(self,data):
    new=node(data)
    if self.head is None:
      self.head=new
    else:
      curr=self.head
      while curr.next is not None:
        curr=curr.next
      curr.next=new
      new.prev=curr
      new.next=None

  def insertAtPosition(self,data,i):
    new=node(data)
    if i==0:
      self.head=new
    else:
      curr=self.head
      for j in range(1,i-1):
        curr=curr.next
      if curr is None:
        print("cant do this operation")
      else:
        k=curr.next
        curr.next=new
        new.prev=curr
        new.next=k
        k.prev=new

  def insertAfterValue(self,data,val):
    new=node(data)
    curr=self.head
    while curr is not None:
      if curr.data==val:
        k=curr.next
        curr.next=new
        new.prev=curr
        new.next=k
        k.prev=new
        if curr.next.prev is not None:
          curr.next.prev=new
        curr.next=new
        break
      curr=curr.next
    if curr is None:
      print("didn't value")

  def deleteFromFront(self):
    if self.head is None:
      print("empty list")
    elif self.head.next is None:
      self.head=None
    else:
      self.head=self.head.next
      self.head.prev=None

  def deleteFromEnd(self):
    if self.head is None:
      print("empty list")
    else:
      curr=self.head
      while curr.next.next is not None:
        curr=curr.next
      curr.next=None

#second approach
  def deleteFromEnd2(self):
    if self.head is None:
      print("empty list")
    else:
      curr=self.head
      before=None
      while curr.next is not None:
        before=curr
        curr=curr.next
      before.next=None

  def deleteAtPos(self,pos):
    if pos==1:
      self.deleteFromFront()
    else:
      curr=self.head
      for i in range(1,pos):
        curr=curr.next
      curr.prev.next=curr.next
      curr.next.prev=curr.prev

  def deleteVal(self,val):
    if self.head is None:
      print("empty list")
    else:
      curr=self.head
      while curr.next is not None:
        if curr.data==val:
          curr.next.prev=curr.prev
          curr.prev.next=curr.next
        curr=curr.next
      if curr is None:
        print("didn't find that value")


  def reverse(self):
    curr=self.head
    prev=None
    while curr is not None:
      next=curr.next
      curr.next=prev
      curr.prev=next
      prev=curr
      curr=next
    self.head=prev






  def printList(self):
    curr=self.head
    while curr.next is not None:
      print(curr.data,end='-')
      curr=curr.next
    print(curr.data)


l=DoublyLinkedList()
l.insertAtHead(10)
l.insertAtHead(20)
l.insertAtHead(30)
l.insertAtHead(40)
print("Insertion at beginning---->")
l.printList()
print("Reversing a linked list---->")
l.reverse()
l.printList()

print("Insertiona at End---->")
l.insertAtEnd(5)
l.printList()
l.insertAtPosition(25,3)
print("Insertion at position---->")
l.printList()
l.insertAfterValue(16,20)
print("Insertion after certain value---->")
l.printList()
l.deleteFromFront()
print("Deletion from Front---->")
l.printList()
l.deleteFromEnd()
print("Deletion from End---->")
l.printList()
l.deleteFromEnd2()
print("Deletion from End2---->")
l.printList()
l.deleteAtPos(2)
print("Deletion at certain position---->")
l.printList()
l.deleteVal(16)
print("Deletion of certain value---->")
l.printList()
```

#Singly Circular Linked List

```python
class node:
  def __init__(self,data):
    self.data=data
    self.next=None
class CircularLinkedList:
  def __init__(self):
    self.head=None
    self.tail=None
  def insertAtFront(self,data):
    new=node(data)
    if self.head is None:
      self.head=new
      self.tail=new
      self.tail.next=self.head
    else:
      new.next=self.head
      self.head=new
      self.tail.next=self.head
  def insertAtEnd(self,data):
    new=node(data)
    if self.head is None:
      self.head=new
      self.tail=new
      self.tail.next=self.head
    else:
      self.tail.next=new
      self.tail=new
      self.tail.next=self.head

  def deleteFromEnd(self):
    if self.head is None:
      print("Empty list")
    else:
      curr=self.head
      while curr.next.next is not self.head:
        curr=curr.next
      curr.next=self.head
      self.tail=curr
      self.tail.next=self.head

  def printList(self):
    curr=self.head
    while curr.next is not self.head:
      print(curr.data,end='-')
      curr=curr.next
    print(curr.data)


l=CircularLinkedList()
l.insertAtFront(10)
l.insertAtFront(20)
l.insertAtFront(30)
l.insertAtFront(40)
print("Insertion From Front--->")
l.printList()
l.insertAtEnd(5)
print("Insertion From End--->")
l.printList()
l.deleteFromEnd()
print("Deletion From End--->")
l.printList()
```

#Implementation of Stack

```python
class ImplementStack:
  def __init__(self,n):
    self.size=n
    self.stack=[None]*n
    self.top=-1

  def push(self,data):
    if self.top==self.size-1:
      print("Cannot add it further")
    else:
      self.top+=1
      self.stack[self.top]=data
  def pop(self):
    if self.top==-1:
      print("cannot remove it further")
    else:
      self.stack[self.top]=None
      self.top-=1

  def peekElement(self):
    if self.top==-1:
      print("empty stack")
    else:
      print(self.stack[self.top])

  def printStack(self):
    for i in range(self.top+1):
      print(self.stack[i],end=" ")
    print()


s=ImplementStack(5)
s.push(10)
s.push(20)
s.push(30)
s.push(40)
print("After pushing--->")
s.printStack()
print("Peek Element after pushing--->")
s.peekElement()
s.pop()
print("After popping--->")
s.printStack()
print("Peek Element after Popping--->")
s.peekElement()





```

```python
stack=[]
def push(data):
  stack.append(data)
def pop():
  if len(stack)==0:
    print("Empty stack")
  else:
    stack.pop()

def peek():
  if len(stack)==0:
    print("Empty stack")
  else:
    print(stack[-1])
push(10)
push(20)
push(30)
push(40)
print("After pushing--->")
print(stack)
print("Peek Element after pushing--->")
peek()
pop()
print("After popping--->")
print(stack)
print("Peek Element after Popping--->")
peek()
```

#Implementation of stack using linked list

```python
class node:
  def __init__(self,data):
    self.data=data
    self.next=None

class LinkedList:
  def __init__(self):
    self.head=None

  def push(self,data):
    new=node(data)
    new.next=self.head
    self.head=new

  def pop(self):
    if self.head is None:
      print("empty stack")
    else:
      curr=self.head
      self.head=curr.next
      print(curr.data)
      curr.next=None

  def peekElement(self):
    if self.head is None:
      print("empty stack")
    else:
      print(self.head.data)

  def printList(self):
    if self.head is None:
      print("empty stack")
    else:
      curr=self.head
      while curr is not None:
        print(curr.data,end=" ")
        curr=curr.next
l=LinkedList()
l.push(10)
l.push(20)
l.push(30)
l.push(40)
print("Top view of stack-->")
l.printList()
print()

print("After Popping top element-->")
l.pop()
print("Peek Element--->")
l.peekElement()




```
